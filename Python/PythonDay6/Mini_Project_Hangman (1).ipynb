```python
import random

words = ["apple", "banana", "orange", "grape", "lemon"]
answer_index = []
answer_Sheets = ""

H = random.choice(words)
print(f"Word to guess: {list(H)}")  # ì„ íƒëœ ë‹¨ì–´ì˜ ê° ë¬¸ìë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ì¶œë ¥

# _ë¡œ ë‹¨ì–´ ê¸¸ì´ì— ë§ê²Œ ì´ˆê¸°í™”
answer_Sheets = "_ " * len(H)
print(f"Initial word: {answer_Sheets}")

# ê° ë¬¸ìë¥¼ _ë¡œ ì´ˆê¸°í™”í•œ ë¦¬ìŠ¤íŠ¸
listed_H = ["_ " for char in H]

# ì‚¬ìš©ì ì…ë ¥ ë°›ê¸°
while "_ " in listed_H:  # _ê°€ ë¦¬ìŠ¤íŠ¸ì— ìˆì„ ë™ì•ˆ ê³„ì† ê²Œì„ ì§„í–‰
    answer = input("Put down Your ANSWER: ")

    # ì…ë ¥í•œ ë¬¸ìê°€ ë‹¨ì–´ì— ìˆìœ¼ë©´
    if answer in H:
        # í•´ë‹¹ ë¬¸ìê°€ ë“±ì¥í•˜ëŠ” ëª¨ë“  ì¸ë±ìŠ¤ ì°¾ê¸°
        answer_index = [i for i, letter in enumerate(H) if letter == answer]

        # í•´ë‹¹ ì¸ë±ìŠ¤ë¥¼ listed_Hì—ì„œ ë°”ê¾¸ê¸°
        for i in answer_index:
            listed_H[i] = answer

        print("Current word: ", ''.join(listed_H))  # í˜„ì¬ê¹Œì§€ ë§íŒ ë‹¨ì–´ ì¶œë ¥

    else:
        print(f"Wrong guess! The letter '{answer}' is not in the word.")

print(f"Congrats! You've guessed the word: {''.join(listed_H)}")
```
