__[과제 4: 관리자 모드 메서드 admin_mode]__

```python
# 리팩토링 과제 힌트
# 구조는 아래와 같습니다.
class BungeoppangShop:

    def __init__(self): 
        self.stock = {
            "Bean": 10,
            "Custard": 8,
            "Choco": 5
        }

        self.prices = {
            "Bean": 1000,
            "Custard": 1200,
            "Choco": 1500
        }

        self.sales = {
            "Bean": 0,
            "Custard": 0,
            "Choco": 0
        }

    def check_stock(self): 
        print(f"####남은 수량####\nBean Flavor: {self.stock["Bean"]} \nCustard Flavor: {self.stock["Custard"]} \nChoco Flavor: {self.stock["Choco"]}")


    def process_order(self, bread_type, bread_count):
        if bread_type not in self.stock:
            print(f'Not exist that flvor.')
            return
        if bread_count > self.stock[bread_type]:
            print("Not enough")

        self.stock -= bread_count
        self.sales += bread_count
        total_price = bread_count * self.prices[bread_type]

        print(f"Total Ammount: {total_price} || You ordered {bread_count} {bread_type} ")


    def admin_mode(self):
        pswd = input("Please Enter Admin Password: ")
        if pswd != "1234":
            print("Access Denied.")
            return

        bread_type = input("Enter bread type to add stock (Bean / Custard / Choco): ")
        if bread_type not in self.stock:
            print("Invalid bread type.")
            return

        try:
            additional_stock = int(input(f"How many {bread_type} to add? "))
            if additional_stock < 0:
                print("You cannot add negative stock.")
                return
        except ValueError:
            print("Invalid input. Please enter a number.")
            return

        self.stock[bread_type] += additional_stock
        print(f"{bread_type} stock updated. New stock: {self.stock[bread_type]}")
                

    def calculate_total_sales(self):
        total = 0
        for bread_type in self.sales:
            total += self.sales[bread_type] * self.prices[bread_type]
        print(f"📊 Total Sales: {total}원")    

def main():
    input(f"\nPARK's 1. STOCK || 2. ORDER || 3. ADMIN || 4. TOTAL SALES")

if __name__ == "__main__":
    main()
```
